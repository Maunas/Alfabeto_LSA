```python
import os
import pandas as pd

metrics_ssd = '/content/drive/MyDrive/Proyecto-Integrador-Redes/metrics-ssd.csv'
metrics_rcnn = '/content/drive/MyDrive/Proyecto-Integrador-Redes/metrics-rcnn.csv'

if os.path.exists(metrics_ssd) and os.path.exists(metrics_rcnn):
    ssd_df = pd.read_csv(metrics_ssd)
    rcnn_df = pd.read_csv(metrics_rcnn)
```

```python
rcnn_df
```

```python
ssd_df
```

```python
import matplotlib.pyplot as plt

plt.plot(ssd_df['Unnamed: 0.1'], ssd_df['test_acc'], label='Test Accuracy SSD')
plt.plot(rcnn_df['Unnamed: 0.4'], rcnn_df['test_acc'], label='Test Accuracy RCNN')

plt.xlabel('Epoch')
plt.ylabel('Value')
plt.title('Test Accuracy por epoch')

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot(ssd_df['Unnamed: 0.1'], ssd_df['train_loss'], label='Train Loss SSD')
plt.plot(rcnn_df['Unnamed: 0.4'], rcnn_df['train_loss'], label='Train Loss RCNN')

plt.xlabel('Epoch')
plt.ylabel('Value')
plt.title('Train Loss por epoch')

plt.legend()

plt.show()
```
